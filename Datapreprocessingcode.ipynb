```python
import pandas as pd
import numpy as np

# Ensure `hea_ECG` is a NumPy array
hea_EMG = np.array(hea_EMG)

# Initialize variables
starttime = 0
fs = 512  # Sampling frequency
labeled_segments = []

# Check for required columns in the `anno` DataFrame
if 'Duration[s]' not in anno.columns or 'Sleep_Stage' not in anno.columns:
    raise ValueError("The 'anno' dataframe must contain 'Duration[s]' and 'Sleep_Stage' columns.")

signaldata = hea_EMG  # Ensure signaldata is a NumPy array

# Process each annotation
for idx, row in anno.iterrows():
    duration = row['Duration[s]']  # Duration in seconds
    segment_length = int(fs * duration)  # Length of the segment in samples
    endtime = starttime + segment_length

    if endtime <= signaldata.shape[0]:  # Check if segment is within signal bounds
        if duration == 30:  # Only process 30-second segments
            # Extract the segment for each channel
            segment = {f'Signal_{i+1}': signal[starttime:endtime].tolist()
                       for i, signal in enumerate(signaldata.T)}  # Transpose for channel-wise slicing
            segment['Sleep_Stage'] = row['Sleep_Stage']  # Add sleep stage
            labeled_segments.append(segment)  # Append segment to the list
        starttime = endtime  # Move to the next segment
    else:
        print(f"Skipping segment {idx}: exceeds signal length.")
        break  # Stop processing further annotations when signal length is exceeded

# Convert labeled segments into a DataFrame
data = pd.DataFrame(labeled_segments)

# Display the resulting DataFrame
print(data)
```

```python
# Find rows where all values in 'Signal_1' are zeros
rows_with_zeros = data.index[data['Signal_1'].apply(lambda x: all(val == 0 for val in x))].tolist()

print(f"Rows with all zeros in 'Signal_1': {rows_with_zeros}")
```

```python
import numpy as np

# Function to reshape a nested list/matrix into a single row
def reshape_matrix_to_row(matrix):
    return np.array(matrix).flatten()  # Convert to numpy array and flatten

# Apply the reshaping function to each element in 'Signal_1'
reshaped_data1 = signal['Signal_1'].apply(lambda x: pd.Series(reshape_matrix_to_row(x)))

# Reset index for the resulting DataFrame
reshaped_data1 = reshaped_data1.reset_index(drop=True)

# Display the reshaped data
reshaped_data1
```

```python
x_bal_1 = reshaped_data1
segment_duration = 2
sampling_rate = 512
segment_length = segment_duration * sampling_rate
segments = []
x_bal_1=np.array(x_bal_1)
for i, signal in enumerate(x_bal_1):
    for start in range(0, len(signal) - segment_length + 1, segment_length):
        segment = signal[start:start + segment_length]
        segments.append(segment)
segments_np = np.array(segments)
two_segments = pd.DataFrame(segments_np)
df1 = pd.concat([two_segments], axis=1)
df1
```
